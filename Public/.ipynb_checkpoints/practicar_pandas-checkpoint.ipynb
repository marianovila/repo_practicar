**CANTIDAD DE EMPLEADOS POR DEPARTAMENTO**
==

```python
#con reset index le asigno un nombre a la colmna del tamaño
df.groupby(['Department']).size().reset_index(name='Cantidad de empleados')
```

**CANTIDAD DE EMPLEADOS POR SALARIO**
==

```python
#con reset index le asigno un nombre a la colmna del tamaño
df.groupby(['salary']).size().reset_index(name='Cantidad')
```

**PROMEDIO DE SATISFACCIÓN SEGÚN SALARIO POR SALARIO**
==

```python
#con reset index le asigno un nombre a la colmna del tamaño
#con sort ordeno los datos

df[['salary','satisfaction_level']].groupby(['salary']).satisfaction_level.mean().sort_values(ascending=False)
```
